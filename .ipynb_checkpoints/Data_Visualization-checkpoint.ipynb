```python
covid_df = pd.read_excel("corona value - 08-04-99.xlsx")
covid_df.columns = covid_df.iloc[1]
covid_df = covid_df.drop(covid_df.index[[0,1,2,3,4]]).dropna().reset_index().drop('index', axis=1)

# print(covid_df.describe(include='all'))
# print(covid_df)
```

```python
cols=[i for i in covid_df.columns if i not in ["Gender"]]
# print(covid_df[cols[52]])
for col in cols:
    covid_df[col]=pd.to_numeric(covid_df[col])

# print(covid_df.Gender)
# print(covid_df.describe())
```

```python
covid_df.to_csv('Covid19_DataSet.csv')
```

```python
from collections import Counter

age_count = Counter(list(covid_df['Age (yr)'].values))
x, y = zip(*age_count.most_common(15))
plt.figure()
plt.barh(x, y)
for i, val in enumerate(y):
    plt.annotate(val, (y[i], x[i]), color='b')
    plt.gca().invert_yaxis()
    plt.ylabel('Brand name')
    plt.xlabel('count')
    plt.grid(False, axis='y')
    plt.show()

```
